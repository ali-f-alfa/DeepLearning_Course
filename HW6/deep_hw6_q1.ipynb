```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-5, 3, 80)
y = x * (1 / (1 + np.exp(-x)))

plt.figure(figsize=(8,4))
plt.plot(x, y)
plt.title('Swish')
plt.grid(True)
plt.axhline(y=0, lw=2, color='k')
plt.axvline(x=0, lw=2, color='k')
plt.show()
```

```python
x = np.linspace(-3, 3, 60)
y = x * np.tanh(np.log(1 + np.exp(x)))

plt.figure(figsize=(6,6))
plt.plot(x, y)
plt.title('Mish')
plt.grid(True)
plt.axhline(y=0, lw=2, color='k')
plt.axvline(x=0, lw=2, color='k')
plt.show()
```

```python
x = np.linspace(-6, 6, 120)
y = (x * np.tanh(np.log(1 + np.exp(x)))) + (1 / (1 + np.exp(-x))) * (1 - (x *np.tanh(np.log(1 + np.exp(x)))))

plt.figure(figsize=(8,6))
plt.plot(x, y)
plt.title('Swish derivatives')
plt.grid(True)
plt.axhline(y=0, lw=2, color='k')
plt.axvline(x=0, lw=2, color='k')
plt.show()
```
